Using the dataset provided, develop the following visualizations using the right
set of variables. (20 each)
a. Histogram
b. Bar chart
c. Line chart
d. Area chart
e. Scatter plot


```python
# Understanding of Data and import of Lib Pandas and matplotlib

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:\data.csv", index_col= 0)
data
```

```python
# a. Histogram
plt.style.use('seaborn-darkgrid')
plt.hist(data['Fuel_Price'])
```

```python
# b. Bar Chart
plt.style.use('seaborn-dark')
data[9:43][['dept_1 sales', 'dept_2 sales']].plot.bar()
data[25:50][['dept_2 sales', 'dept_3 sales']].plot.bar()
data[25:50][['dept_1 sales', 'dept_3 sales']].plot.bar()
```

```python
# c. line Chart
data.plot.line(y='CPI')
plt.style.use('seaborn-dark')
```

```python
# d. area chart
data[["dept_1 sales", "dept_2 sales", "dept_3 sales"]].plot.area()
plt.style.use('seaborn-colorblind')
```

```python
# e. scatter plot

data[0:100].plot.scatter(x='CPI', y='dept_3 sales')
```
